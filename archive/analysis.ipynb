```python
# Import libraries
import pandas as pd
import spacy
from bs4 import BeautifulSoup
import csv
```

```python
# Load job data
df = pd.read_csv("C:\\Users\\chris\\DATA417\\data\\jobs_raw.zip")
df.head()
```

```python
# Get just job descriptions
descriptions = df['Job Description']
len(descriptions)
```

```python
# Function to tidy job description data
def pre_process_text(description, html=True):

    text = description

    if(html):
        soup = BeautifulSoup(description, "html.parser")
        text = soup.get_text(separator=' ')

        # Replace newline characters with spaces
        text = text.replace('\n', ' ')


    # to lower

    # remove URLS

    return text
```

```python
# Tidy and save descriptions
result = descriptions.apply(pre_process_text)
result = result[0:1000]
```
